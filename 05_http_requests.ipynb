```python
from requests import get, post
from bs4 import BeautifulSoup
G_URL = 'https://www.google.ru/search'

def google(query):
    req = get(G_URL, params = {'q' : query})    
    assert req.status_code == 200, 'request failed'
    soup = BeautifulSoup(req.text, "lxml")
    for i, li in enumerate(soup.findAll('div', attrs={'class' : 'kCrYT'})):
        link = li.find('a')
        if link:
            link['href'] = 'http://www.google.com' +  link['href']
            link = '<i>{}.</i>  {}'.format(i, link)
            display(HTML(str(link)))

google(' ')
```
